```python

import pandas as pd
import numpy as np
import networkx as nx
import random
import os

# Directory to save datasets
output_dir = "/mnt/data/blockchain_gnn_datasets"
os.makedirs(output_dir, exist_ok=True)

# 1. Node states: transaction volume, latency, peer connectivity
def calculate_node_states(num_nodes=100):
    data = {
        "node_id": np.arange(num_nodes),
        "transaction_volume": np.random.poisson(lam=20, size=num_nodes),
        "latency": np.random.uniform(10, 100, size=num_nodes),
        "peer_connectivity": np.random.randint(1, 20, size=num_nodes)
    }
    df = pd.DataFrame(data)
    df.to_csv(f"{output_dir}/node_states.csv", index=False)
    return df

# 2. Subgraphs of nodes (shards), node features
def calculate_shard_data(num_nodes=100, num_shards=5):
    nodes = np.arange(num_nodes)
    np.random.shuffle(nodes)
    shard_ids = np.array_split(nodes, num_shards)
    data = []
    for shard_idx, shard in enumerate(shard_ids):
        for node in shard:
            data.append({
                "node_id": node,
                "shard_id": shard_idx,
                "transaction_count": np.random.poisson(15),
                "connectivity": np.random.randint(1, 15)
            })
    df = pd.DataFrame(data)
    df.to_csv(f"{output_dir}/shard_node_features.csv", index=False)
    return df

# 3. Consensus protocol parameters
def calculate_consensus_parameters(num_samples=100):
    data = {
        "sample_id": np.arange(num_samples),
        "block_size": np.random.randint(1, 10, size=num_samples),
        "validation_time": np.random.uniform(0.1, 1.0, size=num_samples),
        "network_latency": np.random.uniform(10, 100, size=num_samples),
        "network_traffic": np.random.randint(100, 1000, size=num_samples)
    }
    df = pd.DataFrame(data)
    df.to_csv(f"{output_dir}/consensus_parameters.csv", index=False)
    return df

# 4. Historical transaction data and interaction patterns
def calculate_transaction_history(num_nodes=100, num_days=30):
    data = []
    for day in range(num_days):
        for node in range(num_nodes):
            data.append({
                "day": day,
                "node_id": node,
                "transactions": np.random.poisson(20),
                "interactions": np.random.randint(1, 25)
            })
    df = pd.DataFrame(data)
    df.to_csv(f"{output_dir}/transaction_history.csv", index=False)
    return df

# 5. Blockchain node graph
def calculate_blockchain_graph(num_nodes=100, edge_prob=0.05):
    G = nx.erdos_renyi_graph(n=num_nodes, p=edge_prob)
    edges = list(G.edges())
    edge_data = [{"source": u, "target": v} for u, v in edges]
    df_edges = pd.DataFrame(edge_data)
    df_edges.to_csv(f"{output_dir}/blockchain_edges.csv", index=False)

    node_data = [{"node_id": node, "transaction_data": np.random.poisson(25)} for node in G.nodes()]
    df_nodes = pd.DataFrame(node_data)
    df_nodes.to_csv(f"{output_dir}/blockchain_nodes.csv", index=False)
    return df_nodes, df_edges

node_states_df = calculate_node_states()
shard_data_df = calculate_shard_data()
consensus_params_df = calculate_consensus_parameters()
transaction_history_df = calculate_transaction_history()
blockchain_nodes_df, blockchain_edges_df = calculate_blockchain_graph()

print('Node States:')
print(node_states_df.head())
print('\nShard Node Features:')
print(shard_data_df.head())
print('\nConsensus Protocol Parameters:')
print(consensus_params_df.head())

!pip install torch_geometric

import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import networkx as nx
import matplotlib.pyplot as plt
from torch_geometric.data import Data
from torch_geometric.nn import GCNConv
from sklearn.preprocessing import MinMaxScaler

# Load node states and graph edges
node_states_df = pd.read_csv("/mnt/data/blockchain_gnn_datasets/node_states.csv")
edges_df = pd.read_csv("/mnt/data/blockchain_gnn_datasets/blockchain_edges.csv")

print("Loaded Node States:")
print(node_states_df.head())

print("\nLoaded Edges:")
print(edges_df.head())

# Normalize node features
features = node_states_df[["transaction_volume", "latency", "peer_connectivity"]].values
scaler = MinMaxScaler()
features = scaler.fit_transform(features)

# Convert edges to PyTorch geometric format
edge_index = torch.tensor(edges_df.values.T, dtype=torch.long)
x = torch.tensor(features, dtype=torch.float)

# Create PyG Data object
data = Data(x=x, edge_index=edge_index)

# Define a simple GNN for policy representation
class GNNPolicy(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim):
        super(GNNPolicy, self).__init__()
        self.conv1 = GCNConv(input_dim, hidden_dim)
        self.relu = nn.ReLU()
        self.conv2 = GCNConv(hidden_dim, output_dim)

    def forward(self, data):
        x, edge_index = data.x, data.edge_index
        x = self.relu(self.conv1(x, edge_index))
        x = torch.sigmoid(self.conv2(x, edge_index))  # Output scaled to [0,1]
        return x

# Initialize model, loss function, optimizer
input_dim = x.shape[1]
hidden_dim = 16
output_dim = 2  # peer selection (0-1), validation timing (0-1)
model = GNNPolicy(input_dim, hidden_dim, output_dim)
optimizer = optim.Adam(model.parameters(), lr=0.01)
loss_fn = nn.MSELoss()

# Simulate training targets (simplified consensus objectives)
np.random.seed(42)
targets = np.random.rand(x.shape[0], output_dim)  # Random desired consensus actions
y = torch.tensor(targets, dtype=torch.float)

# Training loop
print("\nStarting Training Loop...")
for epoch in range(100):
    model.train()
    optimizer.zero_grad()
    out = model(data)
    loss = loss_fn(out, y)
    loss.backward()
    optimizer.step()
    if epoch % 10 == 0:
        print(f"Epoch {epoch}: Loss = {loss.item():.4f}")

# Inference
model.eval()
with torch.no_grad():
    predictions = model(data).numpy()

# Output interpretation
node_ids = node_states_df["node_id"].values
consensus_df = pd.DataFrame({
    "node_id": node_ids,
    "peer_selection_score": predictions[:, 0],
    "validation_timing_score": predictions[:, 1]
})

print("\nOptimal Consensus Strategies per Node:")
print(consensus_df.head())

# Plot results
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(consensus_df["peer_selection_score"], bins=20)
plt.title("Peer Selection Scores")
plt.xlabel("Score")
plt.ylabel("Frequency")

plt.subplot(1, 2, 2)
plt.hist(consensus_df["validation_timing_score"], bins=20)
plt.title("Validation Timing Scores")
plt.xlabel("Score")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()

import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
import networkx as nx
from torch_geometric.data import Data, Batch
from torch_geometric.nn import SAGEConv, global_mean_pool
from sklearn.preprocessing import MinMaxScaler
from collections import defaultdict

# Load the shard data and node states
shard_df = pd.read_csv("/mnt/data/blockchain_gnn_datasets/shard_node_features.csv")
edges_df = pd.read_csv("/mnt/data/blockchain_gnn_datasets/blockchain_edges.csv")

print("Shard Node Features:")
print(shard_df.head())

print("\nEdges:")
print(edges_df.head())

# Normalize features
features = shard_df[["transaction_count", "connectivity"]].values
scaler = MinMaxScaler()
features_scaled = scaler.fit_transform(features)
shard_df[["transaction_count", "connectivity"]] = features_scaled

# Group nodes by shard and prepare subgraphs
shards = defaultdict(list)
for _, row in shard_df.iterrows():
    shards[int(row["shard_id"])].append(int(row["node_id"]))

# Build PyG Data objects for each shard
def build_shard_graph(shard_node_ids, edges_df, features_df):
    node_id_map = {nid: i for i, nid in enumerate(shard_node_ids)}
    shard_edges = edges_df[
        edges_df["source"].isin(shard_node_ids) & edges_df["target"].isin(shard_node_ids)
    ]
    edge_index = shard_edges.applymap(lambda x: node_id_map[x]).values.T
    x = features_df[features_df["node_id"].isin(shard_node_ids)][["transaction_count", "connectivity"]].values
    return Data(
        x=torch.tensor(x, dtype=torch.float),
        edge_index=torch.tensor(edge_index, dtype=torch.long)
    )

shard_graphs = []
for shard_id, node_ids in shards.items():
    shard_graph = build_shard_graph(node_ids, edges_df, shard_df)
    shard_graph.y = torch.tensor([shard_id], dtype=torch.long)
    shard_graphs.append(shard_graph)

# Define GraphSAGE for intra-shard processing
class ShardConsensusModel(nn.Module):
    def __init__(self, input_dim, hidden_dim):
        super(ShardConsensusModel, self).__init__()
        self.sage1 = SAGEConv(input_dim, hidden_dim)
        self.sage2 = SAGEConv(hidden_dim, hidden_dim)

    def forward(self, x, edge_index):
        x = torch.relu(self.sage1(x, edge_index))
        x = self.sage2(x, edge_index)
        return x

# Define D-GNN for inter-shard consensus
class GlobalConsensusModel(nn.Module):
    def __init__(self, input_dim, output_dim):
        super(GlobalConsensusModel, self).__init__()
        self.fc = nn.Sequential(
            nn.Linear(input_dim, 32),
            nn.ReLU(),
            nn.Linear(32, output_dim)
        )

    def forward(self, shard_embeddings, batch):
        pooled = global_mean_pool(shard_embeddings, batch)
        return self.fc(pooled)

# Instantiate models
input_dim = 2
hidden_dim = 16
output_dim = 1
shard_model = ShardConsensusModel(input_dim, hidden_dim)
global_model = GlobalConsensusModel(hidden_dim, output_dim)

# Simulate consensus targets for training
consensus_targets = torch.rand(len(shard_graphs), output_dim)

# Training loop for hierarchical consensus
optimizer = optim.Adam(list(shard_model.parameters()) + list(global_model.parameters()), lr=0.01)
loss_fn = nn.MSELoss()

print("\nStarting Hierarchical Consensus Training...")
for epoch in range(100):
    optimizer.zero_grad()
    shard_embeddings = []
    batch_indices = []

    for i, shard in enumerate(shard_graphs):
        out = shard_model(shard.x, shard.edge_index)
        shard_emb = out.mean(dim=0, keepdim=True)
        shard_embeddings.append(shard_emb)
        batch_indices.append(torch.full((1,), i, dtype=torch.long))

    shard_embeddings = torch.cat(shard_embeddings, dim=0)
    batch = torch.cat(batch_indices)
    consensus_output = global_model(shard_embeddings, batch)
    loss = loss_fn(consensus_output, consensus_targets)
    loss.backward()
    optimizer.step()

    if epoch % 10 == 0:
        print(f"Epoch {epoch}: Loss = {loss.item():.4f}")

# Inference
print("\nComputing Shard and Global Consensus States...")
with torch.no_grad():
    shard_states = []
    for shard in shard_graphs:
        emb = shard_model(shard.x, shard.edge_index).mean(dim=0)
        shard_states.append(emb)
    shard_states = torch.stack(shard_states)
    global_consensus = global_model(shard_states, torch.arange(len(shard_states)))
    consensus_df = pd.DataFrame({
        "shard_id": list(shards.keys()),
        "consensus_state": global_consensus.squeeze().numpy()
    })

print("\nConsensus State of Each Shard:")
print(consensus_df)

# Plot the results
plt.figure(figsize=(8, 5))
plt.bar(consensus_df["shard_id"], consensus_df["consensus_state"], color="skyblue")
plt.title("Shard-Level Consensus States")
plt.xlabel("Shard ID")
plt.ylabel("Consensus State Value")
plt.grid(True)
plt.tight_layout()
plt.show()

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load consensus parameter dataset
consensus_df = pd.read_csv("/mnt/data/blockchain_gnn_datasets/consensus_parameters.csv")
print("Loaded Consensus Parameters Dataset:")
print(consensus_df.head())

# Define the objective function (simulate consensus performance)
# Goal: Maximize throughput and minimize latency and validation time
def objective_function(params):
    block_size, validation_time = params
    if validation_time <= 0:
        return np.inf  # Avoid invalid values
    network_latency = np.mean(consensus_df["network_latency"])
    network_traffic = np.mean(consensus_df["network_traffic"])

    # Simulate performance score (higher is better)
    throughput = (block_size * network_traffic) / (validation_time + network_latency)
    latency_penalty = validation_time + network_latency / block_size
    performance_score = throughput - latency_penalty
    return -performance_score  # Minimize negative performance

# Particle Swarm Optimization (PSO)
class Particle:
    def __init__(self, bounds):
        self.position = np.array([
            np.random.uniform(bounds[0][0], bounds[0][1]),
            np.random.uniform(bounds[1][0], bounds[1][1])
        ])
        self.velocity = np.zeros_like(self.position)
        self.best_position = self.position.copy()
        self.best_score = objective_function(self.position)

    def update(self, global_best, w=0.5, c1=1.0, c2=1.5):
        r1, r2 = np.random.rand(), np.random.rand()
        cognitive = c1 * r1 * (self.best_position - self.position)
        social = c2 * r2 * (global_best - self.position)
        self.velocity = w * self.velocity + cognitive + social
        self.position += self.velocity
        self.position = np.clip(self.position, [1, 0.1], [10, 1.0])  # Respect bounds
        score = objective_function(self.position)
        if score < self.best_score:
            self.best_score = score
            self.best_position = self.position.copy()

# PSO main loop
def pso_optimize(n_particles=30, n_iterations=50, bounds=[(1, 10), (0.1, 1.0)]):
    print("\nStarting Particle Swarm Optimization...")
    swarm = [Particle(bounds) for _ in range(n_particles)]
    global_best = min(swarm, key=lambda p: p.best_score).best_position

    history = []
    for i in range(n_iterations):
        for particle in swarm:
            particle.update(global_best)
        global_best = min(swarm, key=lambda p: p.best_score).best_position
        best_score = objective_function(global_best)
        history.append(-best_score)
        print(f"Iteration {i+1}: Best Score = {-best_score:.4f}, Parameters = Block Size: {global_best[0]:.2f}, Validation Time: {global_best[1]:.2f}")

    return global_best, history

# Run PSO
optimal_params, performance_history = pso_optimize()

# Output final optimized parameters
optimized_block_size, optimized_validation_time = optimal_params
print("\nOptimized Consensus Parameters:")
print(f"Block Size: {optimized_block_size:.2f}")
print(f"Validation Time: {optimized_validation_time:.2f}")

# Plot performance history
plt.figure(figsize=(8, 5))
plt.plot(performance_history, marker='o')
plt.title("PSO Optimization of Consensus Parameters")
plt.xlabel("Iteration")
plt.ylabel("Performance Score")
plt.grid(True)
plt.tight_layout()
plt.show()

import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

# Load transaction history dataset
history_df = pd.read_csv("/mnt/data/blockchain_gnn_datasets/transaction_history.csv")
print("Loaded Transaction History:")
print(history_df.head())

# Aggregate to get average transactions per day (simulate network load)
daily_load = history_df.groupby("day")["transactions"].sum().reset_index()
print("\nAggregated Daily Network Load:")
print(daily_load.head())

# Normalize data
scaler = MinMaxScaler()
load_scaled = scaler.fit_transform(daily_load["transactions"].values.reshape(-1, 1))

# Create sequences for LSTM
def create_sequences(data, window=5):
    sequences, targets = [], []
    for i in range(len(data) - window):
        sequences.append(data[i:i+window])
        targets.append(data[i+window])
    return np.array(sequences), np.array(targets)

window_size = 5
X, y = create_sequences(load_scaled, window=window_size)

# Split into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# Convert to PyTorch tensors
X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train, dtype=torch.float32)
X_test_tensor = torch.tensor(X_test, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test, dtype=torch.float32)

# Define LSTM model
class LoadPredictorLSTM(nn.Module):
    def __init__(self, input_size=1, hidden_size=32, num_layers=1):
        super(LoadPredictorLSTM, self).__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_size, 1)

    def forward(self, x):
        lstm_out, _ = self.lstm(x)
        output = self.fc(lstm_out[:, -1, :])
        return output

# Instantiate model
model = LoadPredictorLSTM()
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)

# Train model
print("\nTraining LSTM for Load Prediction...")
epochs = 100
for epoch in range(epochs):
    model.train()
    optimizer.zero_grad()
    output = model(X_train_tensor)
    loss = criterion(output, y_train_tensor)
    loss.backward()
    optimizer.step()
    if epoch % 10 == 0:
        print(f"Epoch {epoch}: Loss = {loss.item():.4f}")

# Predict
model.eval()
with torch.no_grad():
    predictions = model(X_test_tensor).numpy()
    true_values = y_test_tensor.numpy()

# Inverse scale
predicted_load = scaler.inverse_transform(predictions)
true_load = scaler.inverse_transform(true_values)

# Proactive Adjustment Logic (simulated)
# If predicted load > threshold, increase block size; else decrease validation time
threshold = np.mean(true_load)
adjustments = []
for load in predicted_load:
    if load > threshold:
        adjustments.append("Increase block size")
    else:
        adjustments.append("Decrease validation time")

# Output sample predictions and actions
print("\nPredicted Future Load and Recommended Adjustments:")
for i in range(5):
    print(f"Predicted Load: {predicted_load[i][0]:.2f} â†’ Action: {adjustments[i]}")

# Plot results
plt.figure(figsize=(10, 5))
plt.plot(true_load, label="True Load")
plt.plot(predicted_load, label="Predicted Load", linestyle='--')
plt.title("Network Load Prediction with LSTM")
plt.xlabel("Time Step")
plt.ylabel("Transaction Volume")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import networkx as nx
from sklearn.cluster import SpectralClustering
from sklearn.preprocessing import MinMaxScaler
from torch_geometric.data import Data
from torch_geometric.nn import GCNConv, global_mean_pool

# Step 1: Load data
nodes_df = pd.read_csv("/mnt/data/blockchain_gnn_datasets/blockchain_nodes.csv")
edges_df = pd.read_csv("/mnt/data/blockchain_gnn_datasets/blockchain_edges.csv")
print("Loaded Node and Edge Data")
print(nodes_df.head())
print(edges_df.head())

# Step 2: Build NetworkX graph for spectral clustering
G = nx.from_pandas_edgelist(edges_df, source='source', target='target')
for node in nodes_df['node_id']:
  if node not in G:
    G.add_node(node)
adj_matrix = nx.to_numpy_array(G)
print(f"\nAdjacency matrix shape: {adj_matrix.shape}")

# Step 3: Perform Spectral Clustering
num_shards = 5
clustering = SpectralClustering(n_clusters=num_shards, affinity='precomputed', assign_labels='kmeans')
shard_assignments = clustering.fit_predict(adj_matrix)
nodes_df['shard_id'] = shard_assignments

print("\nShard Assignments:")
print(nodes_df[['node_id', 'shard_id']].head())

# Step 4: Normalize node features
scaler = MinMaxScaler()
features_scaled = scaler.fit_transform(nodes_df[["transaction_data"]])
x = torch.tensor(features_scaled, dtype=torch.float)

# Step 5: Prepare GNN input
edge_index = torch.tensor(edges_df[['source', 'target']].values.T, dtype=torch.long)
data = Data(x=x, edge_index=edge_index)

# Step 6: Define GNN for consensus modeling
class ShardConsensusGNN(nn.Module):
    def __init__(self, in_channels, hidden_channels, out_channels):
        super(ShardConsensusGNN, self).__init__()
        self.conv1 = GCNConv(in_channels, hidden_channels)
        self.conv2 = GCNConv(hidden_channels, out_channels)

    def forward(self, x, edge_index):
        x = torch.relu(self.conv1(x, edge_index))
        return self.conv2(x, edge_index)

# Instantiate GNN
gnn = ShardConsensusGNN(in_channels=1, hidden_channels=8, out_channels=4)
print("\nInitialized GNN model.")

# Step 7: Forward pass and gather shard-wise outputs
gnn.eval()
with torch.no_grad():
    node_embeddings = gnn(data.x, data.edge_index)

# Step 8: Aggregate embeddings per shard for consensus state
shard_states = []
for shard_id in range(num_shards):
    indices = nodes_df[nodes_df['shard_id'] == shard_id].index.tolist()
    shard_embed = node_embeddings[indices]
    shard_state = shard_embed.mean(dim=0)
    shard_states.append(shard_state)

shard_states = torch.stack(shard_states)
global_consensus_state = shard_states.mean(dim=0)

# Step 9: Print and plot results
print("\nShard-wise Consensus States:")
for i, state in enumerate(shard_states):
    print(f"Shard {i}: {state.numpy()}")

print("\nGlobal Inter-Shard Consensus State:")
print(global_consensus_state.numpy())

# Plot shard distribution
plt.figure(figsize=(8, 5))
plt.hist(nodes_df['shard_id'], bins=num_shards, edgecolor='black')
plt.title("Distribution of Nodes Across Shards")
plt.xlabel("Shard ID")
plt.ylabel("Number of Nodes")
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot shard-wise consensus as bar plot
shard_means = [s.mean().item() for s in shard_states]
plt.figure(figsize=(8, 5))
plt.bar(range(num_shards), shard_means, color='skyblue')
plt.title("Shard-Wise Consensus Scores")
plt.xlabel("Shard ID")
plt.ylabel("Consensus Score (Mean Embedding)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
